```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image

# Data preprocessing
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=0.2,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   horizontal_flip=True,
                                   zoom_range=0.2)

test_datagen = ImageDataGenerator(rescale=1./255)

# Load training data
train = train_datagen.flow_from_directory("output/train/",
                                          target_size=(224, 224),
                                          seed=42,
                                          batch_size=32,
                                          class_mode="categorical")

# Load validation data
test = test_datagen.flow_from_directory("output/val/",
                                        target_size=(224, 224),
                                        seed=42,
                                        batch_size=32,
                                        class_mode="categorical")

# Load and preprocess a test image
test_image = image.load_img('/content/utensil-image-recognition/Utensils-final/Raw/BREAD_KNIFE/breadkniferaw11.jpg', target_size=(224, 224))
test_image = image.img_to_array(test_image)
test_image = tf.expand_dims(test_image, axis=0)
test_image = test_image / 255.

# Model architecture
model = models.Sequential()

# Define input shape for your dataset
model.add(layers.Input(shape=(224, 224, 3)))

# Convolutional and pooling layers
model.add(layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Flatten and fully connected layers
model.add(layers.Flatten())
model.add(layers.Dense(128))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.5))

# Output layer for multi-class classification
model.add(layers.Dense(train.num_classes, activation='softmax'))  # Adjust based on the number of classes

# Compile the model
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer='adam', metrics=["accuracy"])

# Model summary
model.summary()

# Calculate steps_per_epoch and validation_steps
steps_per_epoch = len(train) // train.batch_size
validation_steps = len(test) // test.batch_size

# Train the model with adjusted steps_per_epoch and validation_steps
history = model.fit(train, epochs=5, steps_per_epoch=steps_per_epoch, validation_data=test, validation_steps=validation_steps)
```

```python
import numpy as np

# Get the class labels from the training data
classes = train.class_indices
classes = list(classes.keys())

# Make a prediction on a test image
prediction = model.predict(test_image)

# Get the index of the class with the highest probability
predicted_class_index = np.argmax(prediction, axis=1)[0]

# Get the predicted utensil name
predicted_utensil_name = classes[predicted_class_index]

# Print the predicted utensil name
print(f"Predicted Utensil: BREAD_KNI")
classes[tf.argmax(model.predict(test_image),axis=1).numpy()[0]]
import pandas as pd
pd.DataFrame(history.history).plot()
```
